# SEPTIMA DATAFRAME: PRODUCTS.
---
VARIABLES:
>product_id: identificador único de producto. unique product identifier

>product_category_name: categoría raíz del producto en portugués. root category of product, in Portuguese.

>product_name_lenght: número de caracteres extraídos del nombre del producto number of characters extracted from the product name.

>product_description_lenght: Número de caracteres extraídos de la descripción del producto. number of characters extracted from the product description.

>product_photos_qty: número de fotos publicadas del producto. number of product published photos

>product_weight_g: peso del producto medido en gramos. product weight measured in grams.

>product_length_cm: Longitud del producto medida en centímetros. product length measured in centimeters.

>product_height_cm: Altura del producto medida en centímetros. product height measured in centimeters.

>product_width_cm: Ancho del producto medida en centímetros. product width measured in centimeters.

# CREACION de nueva columna PRODUCT_VOLUMEN a partir de:
> product_length_cm,  

>product_height_cm

>product_width_cm


'product_volumen_cm3' Es VOLUMEN en cm3, que se obtiene del PRODUCTO entre Longitud, Altura y Ancho, que vienen expresado en cm.

```python
df7['product_volumen_cm3']= df7['product_length_cm']*df7['product_height_cm']*df7['product_width_cm']
```
